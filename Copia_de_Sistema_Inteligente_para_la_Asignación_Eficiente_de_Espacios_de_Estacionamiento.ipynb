**`Sistema Inteligente para la Asignación Eficiente de Espacios de Estacionamiento`**

```python
import random

# Espacios del estacionamiento
espacios = {
    "P1": {"distancia": 5, "estado": ""},
    "P2": {"distancia": 10, "estado": ""},
    "P3": {"distancia": 15, "estado": ""},
    "P4": {"distancia": 20, "estado": ""},
    "P5": {"distancia": 25, "estado": ""}
}

# Generar estados aleatorios
def generar_estados():
    hay_libre = False

    for espacio in espacios:
        estado = random.choice(["libre", "ocupado"])
        espacios[espacio]["estado"] = estado

        if estado == "libre":
            hay_libre = True

    # Para evitar que todos salgan ocupados
    if not hay_libre:
        espacio_random = random.choice(list(espacios.keys()))
        espacios[espacio_random]["estado"] = "libre"


# Grafo simple del estacionamiento
grafo = {
    "Entrada": {"P1": 5, "P2": 10, "P3": 15, "P4": 20, "P5": 25},
    "P1": {},
    "P2": {},
    "P3": {},
    "P4": {},
    "P5": {}
}


# ALGORITMO VORAZ
def asignar_espacio_voraz():
    mejor_espacio = None
    menor_distancia = 9999

    for espacio, datos in espacios.items():
        if datos["estado"] == "libre":
            if datos["distancia"] < menor_distancia:
                menor_distancia = datos["distancia"]
                mejor_espacio = espacio

    return mejor_espacio


# ALGORITMO DIJKSTRA
def dijkstra(inicio, destino):
    distancias = {}
    visitados = []
    anteriores = {}

    for nodo in grafo:
        distancias[nodo] = 9999
        anteriores[nodo] = None

    distancias[inicio] = 0

    while len(visitados) < len(grafo):
        nodo_actual = None
        menor = 9999

        for nodo in grafo:
            if nodo not in visitados and distancias[nodo] < menor:
                menor = distancias[nodo]
                nodo_actual = nodo

        if nodo_actual is None:
            break

        visitados.append(nodo_actual)

        for vecino, peso in grafo[nodo_actual].items():
            nueva_distancia = distancias[nodo_actual] + peso

            if nueva_distancia < distancias[vecino]:
                distancias[vecino] = nueva_distancia
                anteriores[vecino] = nodo_actual

    camino = []
    actual = destino

    while actual is not None:
        camino.insert(0, actual)
        actual = anteriores[actual]

    return camino, distancias[destino]


# BACKTRACKING
def buscar_alternativa(indice=0, lista_espacios=None):
    if lista_espacios is None:
        lista_espacios = list(espacios.keys())
        random.shuffle(lista_espacios)

    if indice == len(lista_espacios):
        return None

    espacio = lista_espacios[indice]

    if espacios[espacio]["estado"] == "libre":
        return espacio

    return buscar_alternativa(indice + 1, lista_espacios)


# PROGRAMA PRINCIPAL
generar_estados()

print("SISTEMA INTELIGENTE DE ESTACIONAMIENTO")
print("--------------------------------------")

print("\nEstado actual de espacios:")
for espacio, datos in espacios.items():
    print(espacio, "-", datos["estado"], "-", datos["distancia"], "metros")

espacio_asignado = asignar_espacio_voraz()

if espacio_asignado:
    print("\nEspacio asignado por algoritmo Voraz:", espacio_asignado)

    camino, distancia = dijkstra("Entrada", espacio_asignado)

    print("Ruta calculada por Dijkstra:", " -> ".join(camino))
    print("Distancia total:", distancia, "metros")

else:
    print("\nNo hay espacios libres.")


# Simulación de bloqueo del espacio asignado
print("\nSimulación: el espacio asignado se ocupó repentinamente.")

if espacio_asignado:
    espacios[espacio_asignado]["estado"] = "ocupado"

    alternativa = buscar_alternativa()

    if alternativa:
        print("Backtracking encontró una alternativa:", alternativa)

        camino, distancia = dijkstra("Entrada", alternativa)

        print("Nueva ruta:", " -> ".join(camino))
        print("Nueva distancia:", distancia, "metros")
    else:
        print("No hay espacios alternativos disponibles.")
```
