```python
import random

# Grafo del estacionamiento
grafo = {
    "Entrada": {"A": 4, "B": 6},
    "A": {"Entrada": 4, "E1": 3, "E2": 5, "C": 4},
    "B": {"Entrada": 6, "C": 2, "E5": 7},
    "C": {"A": 4, "B": 2, "E3": 3, "E4": 6},
    "E1": {"A": 3},
    "E2": {"A": 5},
    "E3": {"C": 3},
    "E4": {"C": 6},
    "E5": {"B": 7}
}

# Genera espacios libres u ocupados al azar
def generar_espacios_aleatorios():
    espacios = {}

    for i in range(1, 6):
        nombre = "E" + str(i)
        estado = random.choice(["libre", "ocupado"])
        espacios[nombre] = estado

    # Evita que todos salgan ocupados
    hay_libre = False

    for estado in espacios.values():
        if estado == "libre":
            hay_libre = True

    if hay_libre == False:
        espacio_aleatorio = random.choice(list(espacios.keys()))
        espacios[espacio_aleatorio] = "libre"

    return espacios

# Algoritmo Dijkstra
def dijkstra(inicio, destino):
    distancias = {}
    anteriores = {}
    visitados = []

    for nodo in grafo:
        distancias[nodo] = 9999
        anteriores[nodo] = None

    distancias[inicio] = 0

    while len(visitados) < len(grafo):
        nodo_actual = None
        menor_distancia = 9999

        for nodo in grafo:
            if nodo not in visitados and distancias[nodo] < menor_distancia:
                menor_distancia = distancias[nodo]
                nodo_actual = nodo

        if nodo_actual is None:
            break

        visitados.append(nodo_actual)

        for vecino, peso in grafo[nodo_actual].items():
            nueva_distancia = distancias[nodo_actual] + peso

            if nueva_distancia < distancias[vecino]:
                distancias[vecino] = nueva_distancia
                anteriores[vecino] = nodo_actual

    camino = []
    actual = destino

    while actual is not None:
        camino.insert(0, actual)
        actual = anteriores[actual]

    return camino, distancias[destino]


# Algoritmo voraz: elige el espacio libre más cercano
def asignar_espacio_voraz(espacios):
    mejor_espacio = None
    menor_distancia = 9999
    mejor_camino = []

    for espacio, estado in espacios.items():
        if estado == "libre":
            camino, distancia = dijkstra("Entrada", espacio)

            if distancia < menor_distancia:
                menor_distancia = distancia
                mejor_espacio = espacio
                mejor_camino = camino

    return mejor_espacio, mejor_camino, menor_distancia


# Backtracking: busca una alternativa si el espacio asignado se ocupa
def buscar_alternativa(espacios, lista_espacios, indice=0):
    if indice == len(lista_espacios):
        return None

    espacio = lista_espacios[indice]

    if espacios[espacio] == "libre":
        return espacio

    return buscar_alternativa(espacios, lista_espacios, indice + 1)


# Programa principal
espacios = generar_espacios_aleatorios()

print("SISTEMA INTELIGENTE DE ESTACIONAMIENTO")
print("--------------------------------------")

print("\nEstado generado al azar:")
for espacio, estado in espacios.items():
    print(espacio, "-", estado)

espacio_asignado, camino, distancia = asignar_espacio_voraz(espacios)

if espacio_asignado is not None:
    print("\nEspacio asignado:", espacio_asignado)
    print("Ruta más corta:", " -> ".join(camino))
    print("Distancia total:", distancia, "metros")

    print("\nSimulación: el espacio asignado se ocupó repentinamente.")
    espacios[espacio_asignado] = "ocupado"

    lista_espacios = list(espacios.keys())
    alternativa = buscar_alternativa(espacios, lista_espacios)

    if alternativa is not None:
        nuevo_camino, nueva_distancia = dijkstra("Entrada", alternativa)

        print("Nueva alternativa encontrada:", alternativa)
        print("Nueva ruta:", " -> ".join(nuevo_camino))
        print("Nueva distancia:", nueva_distancia, "metros")
    else:
        print("No hay espacios alternativos disponibles.")

else:
    print("\nNo hay espacios libres disponibles.")
```
